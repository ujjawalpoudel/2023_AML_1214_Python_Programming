```python
# Problem 8.3
# T-shirt

# Accept size and message as parameters in function
def make_shirt(size, message):
    print(f'Your t-shirt size is {size} and message is "{message}"')


# Keyword arguments in function call
make_shirt(size="S", message="Hello Nepal")

# Positional arguments in function call
make_shirt("L", "Hello Canada")
```

```python
# Problem 8.4
# Large Shirt


def make_shirt(message, size="L"):
    if size == "L" or size == "M":
        print("I love python")
    else:
        print(message)


# function call when size is L
make_shirt(message="size L", size="L")

# function call when size is M
make_shirt(message="size M", size="M")

# function call when size is XL
make_shirt(message="I love java", size="XL")
```

```python
# Problem 8.7
# Album


def make_album(artist_name, album_name, number_of_tracks=None):
    if number_of_tracks is not None:
        return {
            "artist_name": artist_name,
            "album_name": album_name,
            "number_of_tracks": number_of_tracks,
        }
    else:
        return {
            "artist_name": artist_name,
            "album_name": album_name,
        }


# first function call
print(make_album(artist_name="Michael Jackson", album_name="Off the wall"))

# second function call
print(make_album(artist_name="Taylor Swift", album_name="Anti-Hero"))

# third function call
print(make_album(artist_name="Selena Gomez", album_name="Calm Down"))

# third function call with number of tracks
print(make_album(artist_name="Eminem", album_name="MockingBard", number_of_tracks=5))
```
